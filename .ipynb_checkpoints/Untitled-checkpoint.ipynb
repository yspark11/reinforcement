```python
data = pd.DataFrame(columns = ['아이디', '댓글', '시간'])
for i in range(len(ids)):
    data.loc[i, '아이디'] = ids[i].text.strip()
    data.loc[i, '댓글'] = replies[i].text.strip()
    data.loc[i, '시간'] = pd.to_datetime(times[i].get_attribute('datetime'))
data
```

```python
# 댓글 내용 추출
id_f = []
rp_f = []

# ids  = driver.find_element(By.CSS_SELECTOR, 'div.C4VMK > h3 > div > span > a')
replies = driver.find_element(By.CSS_SELECTOR, 'div._a9zs > span')

for id, reply in zip(ids, replies):
    id_a = id.text.strip()
    id_f.append(id_a)

    rp_a = reply.text.strip()
    rp_f.append(rp_a)


    data = {"아이디": id_f,
            "코멘트": rp_f}

df = pd.DataFrame(data)
df.to_excel('result.xlsx')

driver.quit()
```
